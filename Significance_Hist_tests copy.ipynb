```python
import numpy as np
import healpy as hp
from astropy.io import fits
import matplotlib.pyplot as plt

# Parámetros
GRB = "GRB170803729"
bins = 20
n_seeds = 99
base_path = "/lustre/hawcz01/scratch/userspace/jorgeamontes/GRB_KN/data/healpixz=0_alfa1.5"

def read_significance_map(fname):
    with fits.open(fname) as hdul:
        data = hdul[1].data
        column = 'significance'   # ajusta si el nombre real difiere
        if column not in data.columns.names:
            raise KeyError(f"{column} no encontrado en {fname}")
        return data[column]

all_significances = []

for seed in range(1, n_seeds + 1):
    fname = f"{base_path}/{GRB}/transit_1/{GRB}_transit_1FINAL_C0_FAKE_bkgr_same_dec_pseudo_seed{seed}.fits.gz"
    try:
        sig = read_significance_map(fname)
        sig = sig[sig > -10000]
        all_significances.append(sig)
        print(f"Semilla {seed}: {len(sig)} valores válidos")
    except FileNotFoundError:
        print(f"Archivo no encontrado para seed {seed}, se omite.")
    except Exception as e:
        print(f"Error en seed {seed}: {e}")

if all_significances:
    all_sig = np.concatenate(all_significances)

    # --- Histograma combinado de todas las semillas
    plt.figure(figsize=(8,6))
    plt.hist(all_sig, bins=bins, histtype='step', color='purple')
    plt.xlabel("Significance")
    plt.ylabel("Counts")
    plt.title(f"Histogram of Significance – {GRB} (1–{n_seeds} seeds)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # --- Histogramas individuales y promedio por bin
    # Definimos los bordes de los bins
    bin_edges = np.histogram_bin_edges(all_sig, bins=bins)
    counts_matrix = []

    plt.figure(figsize=(8,6))
    for sig in all_significances:
        counts, _ = np.histogram(sig, bins=bin_edges)
        counts_matrix.append(counts)
        plt.hist(sig, bins=bin_edges, histtype='step', alpha=0.3, color='gray')

    plt.xlabel("Significance")
    plt.ylabel("Counts per seed")
    plt.title(f"Distribuciones individuales – {GRB}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # --- Promedio por bin a través de todas las semillas
    counts_matrix = np.array(counts_matrix)
    mean_counts = counts_matrix.mean(axis=0)
    bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

    plt.figure(figsize=(8,6))
    plt.bar(bin_centers, mean_counts, width=np.diff(bin_edges),
            align='center', color='steelblue', edgecolor='black')
    plt.xlabel("Significance")
    plt.ylabel("Mean counts per bin (across seeds)")
    plt.title(f"Promedio de cuentas por bin – {GRB}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("No se encontraron datos válidos.")
```

# # Para todos los GRBs

```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# --- Configuración
base_path = "/lustre/hawcz01/scratch/userspace/jorgeamontes/GRB_KN/data/healpixz=0_alfa1.5"
bins = 20                  # número de bins del histograma
column_name = "significance"  # ajusta si el nombre real difiere
min_value = -10000         # filtro de valores extremos

def read_significance(fname):
    """Devuelve el array de significancias de un archivo FITS."""
    with fits.open(fname) as hdul:
        data = hdul[1].data
        if column_name not in data.columns.names:
            raise KeyError(f"{column_name} no encontrado en {fname}")
        sig = data[column_name]
    return sig[sig > min_value]

all_significances = []
counts_per_grb = {}   # Para registrar el número de valores válidos por GRB

# --- Recorremos cada subcarpeta de GRB
for grb in sorted(os.listdir(base_path)):
    grb_path = os.path.join(base_path, grb, "transit_1")
    if not os.path.isdir(grb_path):
        continue
    grb_sigs = []

    # buscar todos los archivos de semillas en la carpeta
    pattern = os.path.join(grb_path, f"{grb}_transit_1FINAL_C0_FAKE_bkgr_same_dec_pseudo_seed*.fits.gz")
    for fname in glob.glob(pattern):
        try:
            sig = read_significance(fname)
            grb_sigs.append(sig)
        except Exception as e:
            print(f"Saltando {fname}: {e}")

    if grb_sigs:
        grb_sigs = np.concatenate(grb_sigs)
        all_significances.append(grb_sigs)
        counts_per_grb[grb] = len(grb_sigs)
        print(f"{grb}: {len(grb_sigs)} valores válidos")
    else:
        print(f"{grb}: sin datos válidos")

# --- Histograma total de todos los GRBs
if all_significances:
    all_sig = np.concatenate(all_significances)

    plt.figure(figsize=(8,6))
    plt.hist(all_sig, bins=bins, histtype="step", color="purple")
    plt.xlabel("Significance")
    plt.ylabel("Counts")
    plt.title("Histogram of Significance – Todos los GRBs")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # --- Promedio de cuentas por bin (across all GRBs and seeds)
    bin_edges = np.histogram_bin_edges(all_sig, bins=bins)
    counts_matrix = []

    for grb_sigs in all_significances:
        counts, _ = np.histogram(grb_sigs, bins=bin_edges)
        counts_matrix.append(counts)

    counts_matrix = np.array(counts_matrix)
    mean_counts = counts_matrix.mean(axis=0)
    bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

    plt.figure(figsize=(8,6))
    plt.bar(bin_centers, mean_counts, width=np.diff(bin_edges),
            align="center", color="steelblue", edgecolor="black")
    plt.xlabel("Significance")
    plt.ylabel("Mean counts per bin (across all GRBs)")
    plt.title("Promedio de cuentas por bin – Todos los GRBs")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print("\nResumen por GRB:")
    for k, v in counts_per_grb.items():
        print(f"  {k}: {v} valores válidos")
else:
    print("No se encontraron datos de significancia en ningún GRB.")
```
